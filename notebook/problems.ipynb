## Exercise 1 

Two dices are thrown once and the total score is observed. Use a simulation to find the estimated probability that the total score is even or greater than 7.  A simulation is a repetition of the same experiment multiple times to observe its behavior:

- Run the experiment 1000 times (roll 2 dice 1000 times, and sum the number of both dices).
- Keep track of the number of times that the sum was either greater than 7 or an even number.
- Divide the number from step 2 by the number of iterations (1000).

```python
import numpy as np
import random

def dice_trial():
    dice1 = random.randint(1,6)
    dice2 = random.randint(1,6)
    return dice1+dice2

def sim(n_experiments):
    success = 0
    for i in range(n_experiments):
        result = dice_trial()
        if result > 7 or result%2==0:
            success += 1
    print("Given " +str(n_experiments)+" cases, the probability that two dices"+
          " return a number bigger than 7 or even is "+
          str(round((success / n_experiments)*100,2))+" %")

sim(1000)
```

## Exercise 2

A box contains 10 white balls, 20 red balls and 30 green balls. If we take 5 balls from the box with replacement (we take the ball, observe what color it is and put it back into the box). We want to know the probability of:

1. Take 3 white and 2 red.
2. All are the same color.

Run the experiment 1000 times and calculate the above probabilities.

```python
ball_box = {}

for i in range(60):
    if i < 10:
        ball_box[i] = "White"
    elif (i > 9) and (i < 30):
        ball_box[i] = "Red"
    else:
        ball_box[i] = "Green"

def box_trial():
    result = []
    for x in range(5):
        ball = random.choice(ball_box)
        result.append(ball)
    return result

def sim2(n_experiments): 
    # SET THE CONDITIONS FOR PROBABILITY
    white_red_cond = 0
    all_equal_cond = 0
    for i in range(n_experiments):
        # SET THE VARIABLES OF EACH COLORS
        red_ball = 0
        green_ball = 0
        white_ball = 0    
        result = box_trial()
        #LOOP TO ASSING THE BALLS TO VALUES
        for x in range(len(result)):
            if result[x] == "White":
                white_ball += 1
            elif result[x] == "Green":
                green_ball += 1
            else: red_ball += 1
        # ADD ONE TO THE VALUES IF THE CONDITION IS MET
        if white_ball == 3 and red_ball == 2:
            white_red_cond += 1        
        elif red_ball == 5 or green_ball == 5 or white_ball == 5:
            all_equal_cond += 1
    print("Given "+ str(n_experiments) + " cases, the probabilities are:")
    print("3 white and 2 red - "+ str(round((white_red_cond / n_experiments)*100,2))+" %")
    print("All the same color - "+ str(round((all_equal_cond / n_experiments)*100,2))+" %")
sim2(1000)
```
